### Similarity Measures : Lin, Resnik , WUP, LI, Path, LCH

### Datasets : Verb , Noun , NVA (noun + verb + adjective)

```python
from sematch.semantic.similarity import WordNetSimilarity
from nltk.corpus import wordnet as wn
import pandas as pd
import numpy as np
import glob
from scipy import stats
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr
from scipy.stats import spearmanr
```

```python
wns = WordNetSimilarity()
goldData= glob.glob("datasets/*.csv")
#goldData = ["datasets/NOUN_rw.csv"]
algos = [ 'path','lin', 'res','lch' , 'wup', 'li']#
column_names = ['word1', 'word2', 'sim']
for algoname in algos:
    
    for file in goldData:
        
        print (file)
        ls = []
        word_list = []
        df = pd.read_csv(file)#,  names = column_names) #sep=';',
        # copy word pairs to save result
        df_r = df[['word1', 'word2']]
        # make word pairs
        word_pairs = list(zip(df['word1'], df['word2']))
        if file.split('/')[1].split('_')[0].lower() =='noun':
            print('h')
            for i in word_pairs:
                ls.append(wns.word_similarity(i[0], i[1], algoname, wn.NOUN)) # 0.449327301063
            df_r['sim'] = ls   
            
            df_r.to_csv('./sim_scores/'+algoname+'>'+file.split('/')[1])
        elif file.split('/')[1].split('_')[0].lower() =='verb':
            for i in word_pairs:
                ls.append(wns.word_similarity(i[0], i[1], algoname, wn.VERB)) # 0.449327301063
            df_r['sim'] = ls
            #df_r.to_csv('./sim_scores/'+algoname+'>'+file.split('/')[1])
        else:
            pass
       
```

```python
pearson_ls = []
spearman_ls = [] 
result_dict = {}
goldDatapath= './datasets/'
sim_data = glob.glob("./sim_scores/*.csv")

column_names = ['word1', 'word2', 'sim']
algos = [  'res']
columns =['SimLex222','Agirre201', 'Gerz', 'Millers', 'Rel122', 'wordsim353', 'yang' ]
df_score =pd.DataFrame(columns=columns)
index_ls = []
dfrslt = pd.DataFrame(columns= ['Data','algo','type','Spearman','Pearson'])
    
for file in sim_data:
    print (file)
    algoname1 = file.split('/')[2].split('>')[0]
    
    filename = file.split('/')[-1].split('>')[1]
    # read human similarity score
    df = pd.read_csv(goldDatapath+filename)
    score_human = df['similarity'].values
    norm1 = [(float(i)-min(score_human))/(max(score_human )-min(score_human )) for i in score_human ] # normalize score in 0-1
    
    # read similarity file
    df_r = pd.read_csv(file)
    algo_score = np.round(df_r['sim'].values,2)
    # normalize score
    if (df_r['sim'] > 1).any():
        algo_score= [(float(i)-min(algo_score))/(max(algo_score )-min(algo_score )) for i in algo_score ] # normalize score in 0-1
    
    # calculate Pearson's correlation
    if len(norm1) == len(algo_score):
        corrp, _ = np.round(pearsonr(norm1, algo_score),2)
        corrs, _ = np.round(spearmanr(norm1, algo_score),2)
    
    if (algoname1 == 'lin') or (algoname1 == 'res'):
        IC = 'IC'
    else:
        IC= 'HS'
    dfrslt.loc[len(dfrslt), :]  = [filename,algoname1,IC, corrs,corrp]

#df_score.loc[len(df_score.index)+1] =pearson_ls
    
```
